```python
N, M, K = map(int, input().split())

arr = list(map(int,input().split()))

max_num = 0

next_max_num = 0

answer = 0

arr.sort()

max_num = arr[-1]

next_max_num = arr[-2]

while M:
  for i in range(K):
    if M > 0:
      answer += max_num
    else:
      break
    M -= 1
  if M > 0:
    answer += next_max_num
    M -= 1
  else:
    break
    
print(answer)
```

```python
N, M, K = map(int, input().split())

arr = list(map(int,input().split()))

max_num = 0

next_max_num = 0

answer = 0

arr.sort()

max_num = arr[-1]

next_max_num = arr[-2]

count = (M//(K+1))*K
count += M % (K+1)

answer = count*max_num

answer += (M-count)*next_max_num

print(answer)
```
